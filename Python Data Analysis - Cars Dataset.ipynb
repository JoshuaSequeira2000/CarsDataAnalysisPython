```python
# Importing pandas.

import pandas as pd
```

```python
# Importing cars dataset.

cars = pd.read_csv(r'C:\Users\Joshua\Downloads\Cars.csv')
```

```python
# Viewing the data.

cars.head()
```

```python
# Checking the number of rows and columns in the data.

cars.shape
```

```python
# 1) Finding all null values in the dataset, and filling null values with
#    the mean of its column.

cars.isnull().sum()
```

```python
# 1) a) Replacing '$' in MSRP column with ''

cars['MSRP'] = cars['MSRP'].str.replace('$', '')
cars['MSRP'] = cars['MSRP'].str.replace(',', '')
```

```python
cars.head()
```

```python
# 1) b) Renaming 'MSRP' to 'MSRP (in $)'

cars = cars.rename(columns = {'MSRP': 'MSRP (in $)'})
cars
```

```python
cars.dtypes
```

```python
# 1) c) Converting dtype of MSPR from object to int

cars['MSRP (in $)'] = cars['MSRP (in $)'].astype(float)
```

```python
# 1) d) Filling all the missing values in 'MSRP (in $) with its mean value'

cars['MSRP (in $)'] = cars['MSRP (in $)'].fillna(cars['MSRP (in $)'].mean())
```

```python
cars
```

```python
cars.isnull().sum()
```

```python
# 1) e) Filling the null values of the following columns with the mean of its column.

# EngineSize, Cylinders, Horsepower, MPG_City, MPG_Highway, Weight, Wheelbase, Length.

cars.head()
```

```python
cars['EngineSize'] = cars['EngineSize'].fillna(cars['EngineSize'].mean())
```

```python
cars.isnull().sum()
```

```python
cars['Cylinders'] = cars['Cylinders'].fillna(cars['Cylinders'].mean())
cars['Horsepower'] = cars['Horsepower'].fillna(cars['Horsepower'].mean())
cars['MPG_City'] = cars['MPG_City'].fillna(cars['MPG_City'].mean())
cars['MPG_Highway'] = cars['MPG_Highway'].fillna(cars['MPG_Highway'].mean())
cars['Weight'] = cars['Weight'].fillna(cars['Weight'].mean())
cars['Wheelbase'] = cars['Wheelbase'].fillna(cars['Wheelbase'].mean())
cars['Length'] = cars['Length'].fillna(cars['Length'].mean())
```

```python
cars.isnull().sum()
```

```python
# 1) d) Renaming 'invoice' column to 'invoice (in $)'

cars = cars.rename(columns = {'Invoice': 'Invoice (in $)'})
cars
```

```python
# 1) e) Replacing '$' value in 'Invoice (in $)' to ''

cars['Invoice (in $)'] = cars['Invoice (in $)'].str.replace('$', '')
cars['Invoice (in $)'] = cars['Invoice (in $)'].str.replace(',', '')
cars
```

```python
# 1) f) Converting dtype of cars['Invoice (in $)'] from object to float.

cars['Invoice (in $)'] = cars['Invoice (in $)'].astype(float)
cars
```

```python
cars.dtypes
```

```python
# 1) g) Filling null values in 'Invoice (in $)' with mean values of its column.

cars['Invoice (in $)'] = cars['Invoice (in $)'].fillna(cars['Invoice (in $)'].mean())
cars
```

```python
cars.isnull().sum()
```

```python
cars.dtypes
```
